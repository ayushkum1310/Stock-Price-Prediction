## Description of the dataset
This dataset consists of daily stock prices and volumes of various commodities, cryptocurrencies, and major technology companies. It covers 1,243 data points, each representing a unique date. The features in the dataset include the prices and volumes of natural gas, crude oil, copper, precious metals (platinum, silver, and gold), as well as major tech stocks such as Apple, Tesla, Microsoft, Google, Nvidia, Netflix, Amazon, Meta (Facebook), and Berkshire Hathaway. Additionally, it tracks cryptocurrency prices and volumes, including Bitcoin and Ethereum, along with major market indices like the S&P 500 and Nasdaq 100.It includes:
1. Unnamed: 0: Row index (can be ignored or dropped if not necessary).
2. Date: Date of the stock/commodity data point.
3. Natural_Gas_Price, Natural_Gas_Vol.: Price and volume of natural gas.
4. Crude_oil_Price, Crude_oil_Vol.: Price and volume of crude oil.
5. Copper_Price, Copper_Vol.: Price and volume of copper.
6. Bitcoin_Price, Bitcoin_Vol.: Price and volume of Bitcoin.
7. Platinum_Price, Platinum_Vol.: Price and volume of platinum.
8. Ethereum_Price, Ethereum_Vol.: Price and volume of Ethereum.
9. S&P_500_Price, Nasdaq_100_Price, Nasdaq_100_Vol.: Price of the S&P 500 and Nasdaq 100    indices, and volume for Nasdaq 100.
10. Apple_Price, Apple_Vol.: Price and volume of Apple stocks.
11. Tesla_Price, Tesla_Vol.: Price and volume of Tesla stocks.
12. Microsoft_Price, Microsoft_Vol.: Price and volume of Microsoft stocks.
13. Silver_Price, Silver_Vol.: Price and volume of silver.
14. Google_Price, Google_Vol.: Price and volume of Google stocks.
15. Nvidia_Price, Nvidia_Vol.: Price and volume of Nvidia stocks.
16. Berkshire_Price, Berkshire_Vol.: Price and volume of Berkshire Hathaway stocks.
17. Netflix_Price, Netflix_Vol.: Price and volume of Netflix stocks.
18. Amazon_Price, Amazon_Vol.: Price and volume of Amazon stocks.
19. Meta_Price, Meta_Vol.: Price and volume of Meta (Facebook) stocks.
20. Gold_Price, Gold_Vol.: Price and volume of gold.

## Goal of the Project 
Use historical stock prices and volumes to predict future prices of technology companies (like Apple, Tesla, and Microsoft) and commodities (like oil and gold).

```python
import pandas as pd
import numpy as np
```

## Load the dataset

```python
df=pd.read_csv("Stock Market Dataset.csv")
```

```python
df.info()
```

## Data Preprocessing

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
Missing_column=['Crude_oil_Vol.','Copper_Vol.','Natural_Gas_Vol.','Silver_Vol.','Platinum_Vol.','Gold_Vol.','Nasdaq_100_Vol.']
```

```python
imputer = SimpleImputer(strategy='mean')
df[Missing_column] = imputer.fit_transform(df[Missing_column])
```

```python
df.info()
```

```python
df=df.drop('Date',axis=1)
```

```python
for column in df.columns:
    if df[column].dtype=="object":
        df[column] = df[column].str.replace(',', '').astype(float)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label=LabelEncoder()
```

```python
for column in df.columns:
    if df[column].dtype=="object":
        df[column] = df[column].astype(float)
```

```python
df.info()
```

## Exploratory Data Analysis (EDA)

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Amazon_Price'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Amazon Price')
plt.xlabel('Amazon Price')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

## Split the Data

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['Unnamed: 0','Apple_Price'],axis=1)
y=df['Apple_Price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

## Initialize a Linear Regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

## Model Validation and Testing

```python
from sklearn.model_selection import cross_val_score, KFold
kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
cv_scores = cross_val_score(model, x, y, cv=kf, scoring='r2')  # Using R-squared score
```

```python
mean_score = np.mean(cv_scores)
std_score = np.std(cv_scores)
```

```python
print(f'Cross-validation RÂ² scores: {cv_scores}')
print(f'Mean RÂ² score: {mean_score}')
print(f'Standard deviation of RÂ² score: {std_score}')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
```

```python
import matplotlib.pyplot as plt
```

```python
coefficients = model.coef_
feature_names = x.columns
```

```python
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Coefficient': coefficients
})
```

```python
importance_df['Abs_Coefficient'] = importance_df['Coefficient'].abs()
importance_df = importance_df.sort_values(by='Abs_Coefficient', ascending=False)
```

```python
plt.figure(figsize=(10, 6))
plt.barh(importance_df['Feature'], importance_df['Abs_Coefficient'], color='skyblue')
plt.xlabel('Absolute Coefficient Value (Feature Importance)')
plt.ylabel('Feature')
plt.title('Feature Importances in Linear Regression Model')
plt.gca().invert_yaxis() 
plt.show()
```
